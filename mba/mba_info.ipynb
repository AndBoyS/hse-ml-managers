# Информация о задаче

[Ссылка на датасет](https://docs.google.com/spreadsheets/d/1JiEq5N4QBBkXv7r2JXyOgDRXm5hiI6WE/edit?usp=sharing&ouid=111636018162156614325&rtpof=true&sd=true)

В нашем распоряжении находится датасет, содержащий информацию по клиентам, оформившим ипотечные кредиты. Каждая строка представляет собой данные о конкретном заемщике, включая ключевую информацию о том, произошел ли у этого клиента дефолт на кредит.

Цель - разработать модель машинного обучения, способную предсказывать дефолта клиента на основе имеющихся характеристик и за счет этого максимизировать прибыль банка.

Мы стремимся понять, какие факторы оказывают наибольшее влияние на способность заемщика
выполнять свои обязательства.

Банк всегда выдает кредит под 10% годовых, процент начисляется ежемесячно

### Признаки


- `цель` Указывает на основную цель получения кредита (есть разные типы без названий).
- `срок` Срок кредита в месяцах
- `тип` Определяет категорию кредита (аналогично).
- `тип кредита` Указывает источник данных о кредитном скоре клиента.
- `бизнес` Информирует о том, используется ли кредит для коммерческих целей (Да/Нет).
- `проживание` Описывает тип проживания заемщика – основное, вторичное или инвестиционное.
- `кредитоспособность` Определяет уровень кредитоспособности заемщика, выделяя две категории: более надежные заемщики и менее надежные заемщики.
- `прямой залог` Информирует о наличии прямого залога (конкретного актива) как обеспечения кредита (Да/Нет).
- `амортизация` Указывает на наличие негативной амортизации, что означает, что сумма кредита может возрастать из-за непогашения основной суммы долга.
- `только процент` Тип ипотеки, где некоторое время платится только процент.
- `один платеж` Указывает, производится ли выплата кредита единовременно в конце срока.
- `речь` Отражает характер речи клиента во время получения кредита, что может свидетельствовать о его уверенности и намерениях.
- `лимит_нарушен` Является ли кредит conforming/non_confirming [(справка)](https://www.forbes.com/advisor/mortgages/conforming-vs-non-conforming-loan/)
- `сбор` Слово/словосочетание, косвенно связано с кредитоспособностью
клиента
